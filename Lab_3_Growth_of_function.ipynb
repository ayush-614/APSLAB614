# 1. Growth Of Function
Time Complexity/Order of Growth defines the amount of time taken by any program with respect to the size of the input. Time Complexity specifies how the program would behave as the order of size of input is increased. So, Time Complexity is just a function of size of its input. We will go through some of basic and most common time complexities such as:

- *Constant Time Complexity O(1) : constant running time*
- *Linear Time Complexity O(n) : linear running time*
- *Logarithmic Time Complexity O(log n) : logarithmic running time*
- *Log-Linear Time Complexity O(n log n) : log-Â­linear running time*
- *Polynomial Time Complexity O(n^c) : polynomial running time (c is a constant)*
- *Exponential Time Complexity O(c^n) : exponential running time (c is a constant being raised to a power based on size of input)*

## Constant time Complexity:O(1)
An algorithm is said to have constant time with order O (1) when it is not dependent on the input size n. Irrespective of the input size n, the runtime will always be the same. See example below

```python
# Load the Python libraries
import matplotlib.pyplot as plt
```

```python
# Prob_1

def add_two_numbers(a, b):
    return a + b

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    add_two_numbers(n, n + 1)
    time_values.append(1)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(1) Constant Time Complexity for Addition")
plt.show()
```

```python
import matplotlib.pyplot as plt

def check_even_or_odd(number):
    if number % 2 == 0:
        return "Even"
    else:
        return "Odd"

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    check_even_or_odd(n)
    time_values.append(1)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input (magnitude of number)")
plt.ylabel("Time")
plt.title("O(1) Constant Time Complexity for Even/Odd Check")
plt.show()
```

#Linear Time Complexity O(n) : linear running time
- *An algorithm is said to have a linear time complexity when the running time increases linearly with the length of the input. When the function involves checking all the values in input data, with this order O(n).*

```python
def linear_time_sum(numbers):
    total = 0
    for value in numbers:
        total += value
    return total

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    data = list(range(n))
    linear_time_sum(data)
    time_values.append(n)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(n) Linear Time Complexity")
plt.show()
```

```python
import matplotlib.pyplot as plt

def count_positive_elements(numbers):
    count = 0
    for num in numbers:
        if num > 0:
            count += 1
    return count

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    data = list(range(1, n + 1))
    count_positive_elements(data)
    time_values.append(n)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(n) Linear Time Complexity for Counting Positive Elements")
plt.show()
```

#*Log-Linear Time Complexity O(n log n) : log-Â­linear running time*


* When we call a Logarithmic Time Algorithm inside a loop, it would
result into a Log-Linear Time Complexity program.




```python
#Prob_5 Take Suitable example and write function that runs in log-linear time. plot the graph as well, similar to above example.
```

```python
import matplotlib.pyplot as plt
import math
import random

def merge_sort(arr):
    steps = 0
    if len(arr) > 1:
        mid = len(arr) // 2
        left_half = arr[:mid]
        right_half = arr[mid:]

        steps_left, left_half = merge_sort(left_half)
        steps_right, right_half = merge_sort(right_half)
        steps += steps_left + steps_right

        i = j = k = 0

        while i < len(left_half) and j < len(right_half):
            steps += 1
            if left_half[i] < right_half[j]:
                arr[k] = left_half[i]
                i += 1
            else:
                arr[k] = right_half[j]
                j += 1
            k += 1

        while i < len(left_half):
            steps += 1
            arr[k] = left_half[i]
            i += 1
            k += 1

        while j < len(right_half):
            steps += 1
            arr[k] = right_half[j]
            j += 1
            k += 1
    return steps, arr

n_values = list(range(1, 201, 5))
time_values = []

for n in n_values:

    data = [random.randint(0, 1000) for _ in range(n)]
    total_steps, _ = merge_sort(data.copy())
    time_values.append(total_steps)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time (simulated steps)")
plt.title("O(n log n) Log-Linear Time Complexity (Merge Sort)")
plt.grid(True)
plt.show()
```

### Polynomial Time Complexity O(n^c) : polynomial running time (c is a constant)

* When the computation time increases as function of N raised to some power, N being the size of input. Such a code has Polynomial Time complexity.




```python
# Prob_6 Take Suitable example and write function that runs in log-linear time. plot the graph as well, similar to above example
```

```python
import matplotlib.pyplot as plt

def polynomial_time_pairwise_products(arr):
    steps = 0
    products = []
    for i in range(len(arr)):
        steps += 1
        for j in range(len(arr)):
            steps += 1
            products.append(arr[i] * arr[j])
    return steps, products

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    data = list(range(n))
    total_steps, _ = polynomial_time_pairwise_products(data)
    time_values.append(total_steps)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time (simulated steps)")
plt.title("O(n^2) Polynomial Time Complexity (Pairwise Products)")
plt.grid(True)
plt.show()
```

#Exponential Time Complexity O(c^n) : exponential running time (c is a constant being raised to a power based on size of input)

* When the computation time of a code increases as function of X^N, N being the size of input. Such a code has Polynomial Time Complexity.




```python
# Prob_7 Take Suitable example and write function that runs in log-linear time. plot the graph as well, similar to above example
```

```python
import matplotlib.pyplot as plt

def fibonacci_exponential(n):

    if n <= 1:
        return n, 1
    else:
        fib1, steps1 = fibonacci_exponential(n - 1)
        fib2, steps2 = fibonacci_exponential(n - 2)
        return fib1 + fib2, steps1 + steps2 + 1


n_values = list(range(1, 16))
time_values = []

for n in n_values:
    _, total_steps = fibonacci_exponential(n)
    time_values.append(total_steps)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time (simulated steps)")
plt.title("O(2^n) Exponential Time Complexity (Recursive Fibonacci)")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import math

n_values = list(range(1, 61))

time_o1 = [1 for _ in n_values]
time_ologn = [math.log(n) for n in n_values]
time_on = [n for n in n_values]
time_onlogn = [n * math.log(n) for n in n_values]
time_onsquared = [n**2 for n in n_values]


time_o2n = [2**n for n in n_values]


plt.figure(figsize=(10, 7))
plt.plot(n_values, time_o1, label='O(1) Constant')
plt.plot(n_values, time_ologn, label='O(log n) Logarithmic')
plt.plot(n_values, time_on, label='O(n) Linear')
plt.plot(n_values, time_onlogn, label='O(n log n) Log-Linear')
plt.plot(n_values, time_onsquared, label='O(n^2) Polynomial')


n_values_exp = list(range(1, 21))
time_o2n_subset = [2**n for n in n_values_exp]
plt.plot(n_values_exp, time_o2n_subset, label='O(2^n) Exponential', linestyle='--')

plt.xlabel("Input Size (n)")
plt.ylabel("Operations")
plt.title("Comparison of Time Complexities")
plt.legend()
plt.grid(True)
plt.ylim(0, 5000)
plt.show()
```
