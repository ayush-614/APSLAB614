# Shortest Path Using Dijkstra Algorithm in Python
Problem Description:
Given a weighted graph with non negative edge weights. Find the shortest distance from a source vertex to all other vertices using Dijkstra algorithm

#Theory
Dijkstra algorithm finds shortest distance from one source node to all other nodes in a weighted graph with non negative weights. Dictionary structure stores nodes as keys and neighbor nodes with weights as values.

##Algorithm Steps

1. Represent graph using dictionary.
2. Set distance of every node as infinity.
3. Set distance of source node as 0.
4. Select node with smallest distance among unvisited nodes.
5. Update distance of neighbor nodes.
6. Repeat until all nodes get visited.

```python
def dijkstra(graph, start):

    distances = {}
    visited = []

    for node in graph:
        distances[node] = float('inf')

    distances[start] = 0

    while len(visited) < len(graph):

        min_node = None
        min_distance = float('inf')

        for node in graph:
            if node not in visited and distances[node] < min_distance:
                min_distance = distances[node]
                min_node = node

        visited.append(min_node)

        for neighbor, weight in graph[min_node].items():
            new_distance = distances[min_node] + weight

            if new_distance < distances[neighbor]:
                distances[neighbor] = new_distance

    return distances
```

```python
graph = {
'A': {'B': 4, 'C': 2},
'B': {'A': 4, 'C': 5, 'D': 10},
'C': {'A': 2, 'B': 5, 'D': 3},
'D': {'B': 10, 'C': 3}
}

result = dijkstra(graph, 'A')

for node in result:
    print("Distance from A to", node, "=", result[node])
```

```python
def dijkstra(graph, start):
    distances = {node: float('inf') for node in graph}
    predecessors = {node: None for node in graph}
    distances[start] = 0
    visited = set()

    while len(visited) < len(graph):
        min_node = None
        for node in graph:
            if node not in visited:
                if min_node is None or distances[node] < distances[min_node]:
                    min_node = node

        if min_node is None or distances[min_node] == float('inf'):
            break

        visited.add(min_node)

        for neighbor, weight in graph[min_node].items():
            if neighbor in distances:
                new_distance = distances[min_node] + weight
                if new_distance < distances[neighbor]:
                    distances[neighbor] = new_distance
                    predecessors[neighbor] = min_node

    return distances, predecessors

def get_path(predecessors, start, target):
    path = []
    current = target
    while current is not None:
        path.append(current)
        current = predecessors[current]
    path.reverse()
    return path if path and path[0] == start else []

graph = {}
try:
    num_nodes = int(input("Enter number of total nodes: "))
    node_names = []
    for i in range(num_nodes):
        name = input(f"Enter name for node {i+1}: ").strip()
        node_names.append(name)
        graph[name] = {}

    print("\n--- Define Edges ---")
    print("Format: neighbor:weight (e.g., B:4 C:2)")
    for node in node_names:
        entry = input(f"Enter neighbors for {node}: ").strip()
        if entry:
            connections = entry.split()
            for item in connections:
                neighbor, weight = item.split(':')
                graph[node][neighbor] = int(weight)
except Exception as e:
    print(f"Input error: {e}")
    exit()

source_node = input("\nEnter source node: ").strip()

if source_node in graph:
    distances, predecessors = dijkstra(graph, source_node)
    print(f"\nResults from {source_node}:")
    for target in graph:
        path = get_path(predecessors, source_node, target)
        path_str = " -> ".join(path)
        dist = distances[target]
        print(f"To {target}: Distance = {dist}, Path = [{path_str}]")
else:
    print("Source node not found.")
```
